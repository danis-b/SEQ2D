```python
import numpy as np
from scipy.sparse.linalg import eigsh
from scipy.sparse.linalg import eigs
from scipy import sparse
import matplotlib.pylab as plt


# def rad(i,n):
#     #vectors in angstom
#     a1 = np.array([5,0])
#     a2 = np.array([0,5])
    
#     m = int(n/2)
    
#     x = (int(i % n) - m) 
#     y = (int(i / n) - m)
    
#     r = a1*(x/m) + a2*(y/m)
    
#     return np.linalg.norm(r)
    

N = 80 
Z = 0.2 # effective charge
m = 1   # effective mass

X,Y = np.meshgrid(np.linspace(-2, 2, N, dtype=float), np.linspace(-2, 2, N, dtype=float))

def get_potential(x,y):
    return 0*x

V  = get_potential(X,Y)

diag = np.ones([N])
diags = np.array([diag, -2*diag, diag])
D = sparse.spdiags(diags, np.array([-1,0,1]),N,N)
T = -1/2*sparse.kronsum(D,D)
U = sparse.diags(V.reshape(N**2), (0))
H = T+U

eigenvalues, eigenvectors = eigsh(H, k = 5, which = 'SM')

def get_e(n):
    return eigenvectors.T[n].reshape((N,N))


plt.figure(figsize=(5,5))
plt.contour(X,Y,get_e(0)**2, 100)
```
